```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from gensim.models.ldamodel import LdaModel
from gensim.corpora.dictionary import Dictionary
```

```python
def get_stopword_list():
    stop_word_path = './stopwords.txt'
    stopword_list = [sw.replace('\n', '') for sw in open(stop_word_path,encoding='utf8').readlines()]
    stop_words = set(stopword_list) 
    return stop_words

stop_words = set(stopwords.words('english'))
ex_stop = [
    'The',
    'product','products',
    'users',
    'design',
    '.',
    "'s",
    ':',
    ',', '.', ';', ':', "'", '"', '?', '!', '-', '_', '(', ')', '[', ']', '{', '}',
    'use','also','used','time','needs',"''","''","'s","It","This","this","``","function","make","people",'0','1','2','3','4','new','user'
]
stop_words.update(ex_stop)
stop_words.update(get_stopword_list())
```

```python
# tokenize the text
finally_text["token"] = finally_text["text"].apply(lambda x: word_tokenize(x))

#remove stop words
finally_text["token"] = finally_text["token"].apply(lambda x: [item for item in x if item not in stop_words])

#create a dictionary from token
dictionary = Dictionary(finally_text["token"])


#convert the corpus to a bag of words representation
corpus = [dictionary.doc2bow(text) for text in finally_text['token']]

# Train the LDA model
lda_model = LdaModel(corpus, num_topics=10, id2word=dictionary)

```

```python
# Print the topics and their respective keywords
for idx, topic in lda_model.print_topics():
    print(topic)
```

```python
import gensim
from gensim import corpora
df = finally_text
# Load the text data from DataFrame
text_data = list(df['text'].values)

# Tokenize the text data
tokenized_text = [doc.split() for doc in text_data]

# Create a dictionary from the tokenized text data
dictionary = corpora.Dictionary(tokenized_text)

# Convert tokenized text into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in tokenized_text]

# Build the LDA model
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=10)

# Extract keywords from the LDA model for each document
keywords_list = []
for doc in corpus:
    keywords = lda_model[doc]
    keywords = [word[0] for word in keywords]
    keywords = [dictionary[word] for word in keywords]
    keywords_list.append(" ".join(keywords))

# Add keywords column to DataFrame
df['keywords'] = keywords_list

# Save the DataFrame to csv
df.to_csv('output.csv', index=False)
```

#TF_IDF

```python
import pandas as pd
import gensim
from gensim import corpora
from gensim.models import TfidfModel
from gensim.parsing.preprocessing import remove_stopwords
from nltk.stem import SnowballStemmer
df = finally_text
# Load the stop words
with open('stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.read().splitlines()

# Create a stemmer
stemmer = SnowballStemmer('english')

# Define a function to preprocess the text data
def preprocess(text):
    # Remove stop words
    text = remove_stopwords(text)
    # Tokenize the text
    tokens = gensim.utils.simple_preprocess(text)
    # Stem the tokens
    tokens = [stemmer.stem(token) for token in tokens]
    return tokens

# Load the text data from DataFrame
text_data = list(df['text'].values)

# Preprocess the text data
processed_text = [preprocess(text) for text in text_data]

# Create a dictionary from the tokenized text data
dictionary = corpora.Dictionary(processed_text)

# Convert tokenized text into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in processed_text]

# Build the TF-IDF model
tfidf = TfidfModel(corpus)

# Extract keywords from the TF-IDF model for each document
keywords_list = []
for doc in corpus:
    tfidf_weights = tfidf[doc]
    sorted_tfidf_weights = sorted(tfidf_weights, key=lambda w: w[1], reverse=True)
    top_n = min(len(sorted_tfidf_weights), 20)
    top_n_weights = sorted_tfidf_weights[:top_n]
    top_n_weights = [weight[0] for weight in top_n_weights]
    top_n_tokens = [dictionary[token] for token in top_n_weights]
    keywords_list.append(" ".join(top_n_tokens))

# Add keywords column to DataFrame
df['keywords'] = keywords_list

# Save the DataFrame to csv
df.to_csv('output.csv', index=False)
```
